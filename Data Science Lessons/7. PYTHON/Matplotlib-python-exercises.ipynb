# Matplotlib in python - for visualization 

## line charts

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 10)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("time")
plt.ylabel("function of time")
plt.title("my cool chart")
```

```python
x = np.linspace(0,10,100) # try with more points for smoother graph
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel("time")
plt.ylabel("function of time")
plt.title("my cool chart")
```

## Plotting images in python
- get the data from this website to visualize: https://www.kaggle.com/loovmj/first-mnist-challenge/data

```python
# train.csv data is large, need to download from kaggle site
df = pd.read_csv("train.csv")
df.shape
M = df.as_matrix()
im = M[0,1:]
im.shape

# want 28x28 matrix for an image so re-shape
im = im.reshape(28,28)
plt.imshow(im)
plt.show()
M[0,0]

# set color map
plt.imshow(im, cmap = "gray")
plt.show()

# another option for color
#255 is the maximum number in a byte to make the image negative
plt.imshow(255 - im, cmap = "gray") 
```
